Importing essential libraries.

```python
from ucimlrepo import fetch_ucirepo
import pandas
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

We will store the features and targets. And then we will cut the last 1000 line from the data to use it on trying the decision tree.

```python
# fetch dataset
bike_sharing = fetch_ucirepo(id=275)

# data (as pandas dataframes)
X = bike_sharing.data.features
y = bike_sharing.data.targets

X = X.drop(columns=['dteday'])

trial_X = X[-1000:] # keep last 1000 line for trial
trial_y = y[-1000:]

X = X[:-1000] # remove last 1000 from training data
y = y[:-1000]

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X,y)
```

Decision tree is ready to predict.

```python
predictions = dtree.predict(trial_X)
# Reset the index of the trial_y DataFrame to start from 0
trial_y = trial_y.reset_index(drop=True)
trial_y = trial_y.values

error = []
print('Prediction', "\t", 'Actual Value', "\t", 'Error')
for i in range(predictions.size):
  error_percentage = ((predictions[i] - trial_y[i]) / trial_y[i]) * 100
  error.append(error_percentage)
  print(predictions[i], "\t\t", trial_y[i], "\t\t", error_percentage)
```

```python
print("\nAverage error percentage = ", sum(error) / len(error))
```

As you can see, the error percentage changes around %50. So that leaves us the accuracy of this decision tree is around %50 too.

Here is the set of rules of the decision tree.

```python
text_representation = tree.export_text(dtree,  feature_names=X.columns.tolist())
print(text_representation)
```
