Here is the main pca function. It takes a matrix and number of the components.

It calculates the mean , eigenvalues and eigenvectors with the help of covariance matrix.

It sorts the eigenvalues and eigenvectors. Then selects the first "n_components" principal components. In this homework's case, i need 4 of them.

Then it returns mean, weights (eigenvalues) and vectors (PC#s)

**PART 1:**

```python
import numpy as np

def pca(X, n_components=4):
    # Compute the mean of the columns of X
    mean = np.mean(X, axis=0)

    # Center the data
    X_centered = X - mean

    # Compute the covariance matrix
    cov_matrix = np.cov(X_centered, rowvar=False)

    # Compute eigenvalues and eigenvectors of the covariance matrix
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

    # Sort eigenvalues and corresponding eigenvectors
    sorted_indices = np.argsort(eigenvalues)[::-1]
    sorted_eigenvalues = eigenvalues[sorted_indices]
    sorted_eigenvectors = eigenvectors[:, sorted_indices]

    # Select the first 4 principal components
    principal_components = sorted_eigenvectors[:, :n_components]
    eigenvalues = sorted_eigenvalues[:4]

    return mean, eigenvalues, principal_components
```

Covariance measures how two features vary with each other. A positive covariance indicates that features increase and decrease together.

The eigenvectors represent the principal components (the directions of maximum variance) of the covariance matrix. The eigenvalues are their corresponding magnitude. The eigenvector that has the largest corresponding eigenvalue represents the direction of maximum variance. (Principal Component 1)

And the principal component that is perpendicular to PC1 is called PC2.

```python
# Choose 1000 random items
random_indices = np.random.choice(train_X.shape[0], 1000, replace=False)
random_images = train_X[random_indices]
random_targets = train_y[random_indices]
scaled = random_images / 255.0
```

Picked 1000 data and scaled it between 0 and 1 to normalize the data.

```python
mean, weights, vectors = pca(scaled[0])

# print("\nFirst 4 Eigenvectors (principal components):")
# print(vectors[:, 0])

dot_product = np.dot(vectors[:, 0], vectors[:, 1])
print("Dot product between PC1 and PC2:", dot_product)
# This should give a result that is very close to 0. Becuase PC1 and PC2 are perpendicular to each other.
```

Now, we are going to flatten the images to treat each pixel as a feature. That way PCA will be easier and more accurate.

And plotting the PC1 vs PC2 and PC1 vs PC3 to show some clustering of the labels.

```python
flattened_images = scaled.reshape(scaled.shape[0], -1)

# Apply PCA to find the first and second and third principal components
mean, eigenvalues, eigenvectors = pca(flattened_images)

PC1 = eigenvectors[:, 0]
PC2 = eigenvectors[:, 1]
PC3 = eigenvectors[:, 2]

# Project the images onto the first and second principal components
projected_images_pc1_pc2 = np.dot(flattened_images - mean, eigenvectors[:, [0, 1]])

# Project the images onto the first and third principal components
projected_images_pc1_pc3 = np.dot(flattened_images - mean, eigenvectors[:, [0, 2]])

fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# PC1 vs PC2
axs[0].scatter(projected_images_pc1_pc2[:, 0], projected_images_pc1_pc2[:, 1], s=5)
axs[0].set_xlabel('PC1')
axs[0].set_ylabel('PC2')
axs[0].set_title('PC1 vs PC2')
axs[0].grid(True)

# PC1 vs PC3
axs[1].scatter(projected_images_pc1_pc3[:, 0], projected_images_pc1_pc3[:, 1], s=5)
axs[1].set_xlabel('PC1')
axs[1].set_ylabel('PC3')
axs[1].set_title('PC1 vs PC3')
axs[1].grid(True)

plt.show()
```

```python
projected_images_pc1_pc2_real = np.abs(projected_images_pc1_pc2)

rf = RandomForestClassifier(n_estimators=100, random_state=42)

rf.fit(projected_images_pc1_pc2_real, random_targets)

test_X_flattened = test_X.reshape(test_X.shape[0], -1)
projected_test_data = np.dot(test_X_flattened - mean, eigenvectors[:, :2])
projected_test_data_real = np.abs(projected_test_data)

# Predict on the test set
y_pred = rf.predict(projected_test_data_real)

accuracy = accuracy_score(test_y, y_pred)
print("Accuracy:", accuracy)
```
